```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(stop_words='english', max_features=2000)
```

```python
def listToString(L):
    str = ','.join(L)
    str = str.replace(',', ' ')
    str = str.lower()
    return str
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
def stem(text):
    return ' '.join([stemmer.stem(word) for word in text.split()])
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
movies['tags'][5]
```

```python
count_matrix = vectorizer.fit_transform(movies['tags'])
```

```python
vectorArray = count_matrix.toarray()
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix_array = cosine_similarity(vectorArray)
```

```python
similarity_matrix_array[0]
```

```python
def recommend(movie_name):
    index = movies[movies['title']==movie_name].index[0]
    distances = similarity_matrix_array[index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]
    for i in movies_list:
        print(movies['title'][i[0]])
```

```python
recommend('The Shawshank Redemption')
```

```python
recommend('The Secret Life of Pets')
```

```python
recommend('Last Action Hero')
```
